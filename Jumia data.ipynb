## we want to fetch the list of phones available on Jumia Store catalogue and put the obtained data in a csv file ##

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

We are going to fetch the gadgets listed in tthe catalogue and their attributes such as name, pricing and rating

```python
current_page = 1
data = []
proceed = True 


while (proceed):
    
    url = "https://www.jumia.co.ke/phones-tablets/?page-"+str(current_page)+"#catalog-listing"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    if soup.title.text == "404 Not Found":
        proceed = False
    else:
        gadgets = soup.find_all("div", class_ ="info")

    for gadget in gadgets:
        item = {}
        item["name"] = gadget.find("h3", class_ = "name")
        item["price"] = gadget.find("h3", class_ = "prc")
        item["rating"] = gadget.find("div", class_ = "stars _s")
        print(item["name"])
        data.append(item)
    current_page +=1
```

Putting the data in a csv file

```python
jumia_phones = Pd.DataFrame(item)
jumia_phones.to_csv("Jumia_phones.csv", index = False)
```
